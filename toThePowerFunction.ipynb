```python

####  Method 1  ->>>   Simple

def toThePower(x:int = None, y:int = None) ->int:
  """   To the power, simple
          also check,  2 to the power 1000  and time taken for it
  """
  import time
  start =  time.time()
  if y < 0 or x == 0:
    return 0

  product =  1

  for i in range(y):
    product *= x

  end = time.time()
  time_taken =  end - start


  return (product, time_taken)


x = 10000
y = 1003


result, time_taken =  toThePower(x,y)

### 1 second   =  1000 milliseconds

print(f" {x}  to the power of {y}   =    {result}")
print(f" \n\n {x}  to the power of {y}  time taken is  =    {time_taken * 1000}    milliseconds ")
```

```python

##### Meythod  2    ->    When base is  10

def toThePowerTimeefficientWay(x:int = None, y:int = None) ->int:
  """   To the power, to save little time
          also check,  2 to the power 1000  and time taken for it
  """
  import time

  start =  time.time()


  if y < 0 or x == 0:
    return 0

  product =  1

  for i in range(y//10):
    product *= x*x*x*x*x*x*x*x*x*x

  if y%10 == 0:
    pass
  elif y%10 ==9:
    product *= x
  elif y%10 ==8:
    product *=  x*x
  elif y%10 ==7:
    product *=   x*x*x
  elif y%10 ==6:
    product *= x*x*x*x
  elif y%10 ==5:
    product *= x*x*x*x*x
  elif y%10 ==4:
    product *= x*x*x*x*x    *x
  elif y%10 ==3:
    product *= x*x*x*x*x    *x*x
  elif y%10 ==2:
    product *= x*x*x*x*x     *x*x*x
  elif y%10 ==1:
    product *= x*x*x*x*x      *x*x*x*x

  end = time.time()
  time_taken =  end - start


  return (product, time_taken)


x = 10000
y = 1003


result, time_taken =  toThePowerTimeefficientWay(x,y)

### 1 second = 1000  milliseconds

print(f" {x}  to the power of {y}   =    {result}")
print(f" \n\n {x}  to the power of {y}  time taken is  =    {time_taken* 1000}   milliseconds   ")
```

```python
def compareTime(method_simple_time = None, method_base_10_time  = None) ->float:
  """
       This is function to comapre time taken
       (difference) b/w execution done
       by differentt method
       in milliseconds
  """
  diff = method_simple_time - method_base_10_time
  percetage = (method_simple_time / method_base_10_time ) * 100
  return percetage




method_simple_time =  0.7059574127197266
method_base_10_time  = 0.2849102020263672

percentage = compareTime(method_simple_time,  method_base_10_time)


print(f" method_simple_time  is {method_simple_time}   and method_base_10_time is {method_base_10_time}   for same input in same sys() at same time\n")
print(f" the difference is  method_simple_time()/method_base_10_time()  in percent is  {percentage} \n")
###print(f"if result in minus  that means method_simple_time() is faster by given percentage")
print(f"if result in plus  that means method_base_10_time() is  {percentage}  time faster to method_simple_time() ")
```

```python
############################   Now recursive   ###############

import time

def tothePowUsingRec(x = None, y = None) ->tuple:
  """
      To the Power using recursive method
      will return prduct & time taken to calculate this

   """
  import time
  global start
  start = time.time()
  product = 1

  if y == 0:
    return 1

  if y%2 == 0:
    product = tothePowUsingRec(x, y//2)
    return product * product
  else:
    product = tothePowUsingRec(x, (y-1)//2)
    return product * product * x   ###  because y is 1 less    eg 11 -> 11/2 --> (5 * 5) * ek baaar aur with x



################################ main ########################
x = 10000
y = 1003

global start
result =  tothePowUsingRec(x,y)
end =  time.time()
time_taken = end -start


print(f" {x}  to the power of {y}   =    {result}")
print(f" \n\n {x}  to the power of {y}  time taken is  =    {time_taken* 1000}   milliseconds   ")
```
